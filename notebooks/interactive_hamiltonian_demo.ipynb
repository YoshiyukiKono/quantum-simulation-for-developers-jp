# Interactive Hamiltonian Simulation Demo

このNotebookでは、1量子ビットHamiltonian

$$
H = Z
$$

に対する時間発展

$$
U(t) = e^{-iHt}
$$

をスライダーで動かしながら可視化します。

## 1. ライブラリの読み込み

```python
import numpy as np
import matplotlib.pyplot as plt

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, SparsePauliOp
from qiskit.circuit.library import PauliEvolutionGate
from qiskit.visualization import plot_bloch_multivector

import ipywidgets as widgets
from IPython.display import display
```

## 2. 時間発展回路を作る関数

ここでは、初期状態を Hadamard ゲートで $|+\rangle$ にしてから、

$$
U(t) = e^{-iZt}
$$

を適用します。

```python
def create_time_evolution_circuit(t):
    hamiltonian = SparsePauliOp(["Z"], coeffs=[1.0])
    evolution_gate = PauliEvolutionGate(hamiltonian, time=t)

    qc = QuantumCircuit(1)
    qc.h(0)
    qc.append(evolution_gate, [0])

    return qc
```

## 3. 状態ベクトルと回路を表示する

```python
def show_state(t):
    qc = create_time_evolution_circuit(t)
    state = Statevector.from_instruction(qc)

    print("time t =", round(t, 3))
    print(state)

    display(qc.draw("mpl"))
```

## 4. スライダーで時間発展を動かす

```python
widgets.interact(
    show_state,
    t=widgets.FloatSlider(
        value=0.0,
        min=0.0,
        max=2 * np.pi,
        step=0.1,
        description="time t"
    )
);
```

## 5. Bloch球で時間発展を見る

$H = Z$ による時間発展は、Bloch球上ではZ軸まわりの回転として見えます。

```python
def show_bloch(t):
    qc = create_time_evolution_circuit(t)
    state = Statevector.from_instruction(qc)

    print("time t =", round(t, 3))
    display(plot_bloch_multivector(state))
```

```python
widgets.interact(
    show_bloch,
    t=widgets.FloatSlider(
        value=0.0,
        min=0.0,
        max=2 * np.pi,
        step=0.1,
        description="time t"
    )
);
```
